Matplotlib è una potente libreria che permette di realizzare facilmente grafici. Per utilizzarla iniziamo importando i tool necessari:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Ci sono alcuni elementi base che dovremo utilizzare per realizzare un grafico. Vediamoli nel dettaglio 

## Figure

Il primo è `figure` che è il "contenitore" del nostro grafico.
Iniziamo quindi creando una figura a cui assegneremo un titolo:

```python
fig = plt.figure()
fig.suptitle('Questa figura non ha assi!')
```

La gestione dell'impaginazione in matplotlib è molto semplice. Se vogliamo, ad esempio, creare 4 grafici disposti a griglia nella stessa figura ci basta usare il comando `subplots` aggiungendo alla nostra figura anche gli assi necessari.

```python
fig, ax_lst = plt.subplots(2, 2)
```

## Assi

Gli assi sono gli oggetti che delimitano i grafici e definiscono lo spazio in cui sono contenuti i dati. Un grafico bidimensionale ne ha *almeno* due, mentre uno tridimensionale ne avrà *almeno* tre.

Come detto gli assi definiscono i **limiti** di visualizzazione dei nostri dati e tali valori possono essere modificati con i metodi `set_xlim()` e `set_ylim`. 

Si possono cambiare i **nomi degli assi** tramite `set_xlabel` e `set_ylabel`.

Per definire l'intervallo delle tacche degli assi possiamo usare il metodo `xtixcks` o `yticks`, assegnando o meno le etichette per ogni tacca. Lo vedremo meglio quando realizzeremo il primo grafico. (vedi https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xticks.html)

Possiamo usare molte altre opzioni per migliorare la visualizzazione del nostro grafico come la dimensione del carattere delle etichette degli assi o delle tacche.

## Il primo grafico

```python
# creiamo un array di 100 valori equispaziati fra 0 e 2
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='lineare')
plt.plot(x, x**2, label='quadratica')
plt.plot(x, x**3, label='cubica')

plt.xlabel('x label', fontsize="14")
plt.ylabel('y label', fontsize="14")

plt.title("Plot semplice", fontsize="14")

plt.tick_params(axis='both',labelsize=14)

plt.legend()

plt.show()
```

Un tipo di grafico molto comune è quello composto da punti scatter in cui i dati vengono individuati da una coppia di coordinate:

```python
plt.scatter(2,4)
plt.show()
```

Se abbiamo due array di dati possiamo utilizzarli come coordinate per i punti da plottare:

```python
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.scatter(x_values, y_values)

plt.show()
```
